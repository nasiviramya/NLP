# POSTERSTEMMER

```python
words = ["running", "happily", "studies", "caring", "better", "easily", "jumps", "playing", "writing", "leaves"]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + "----->" + stemming.stem(word))
```

* stemming doesn’t always return perfect words because it follows rules rather than meaning
*  words like "better" will still stay the same
*  But, stemming doesn’t handle irregular cases well.
 ** ex:  happily----->happili
         leaves----->leav

# REGEXPSTEMMER CLASS

```python
from nltk.stem import RegexpStemmer  
```

```python
reg = RegexpStemmer('ing$|s$|es$|able$', min=4)
```

```python
reg.stem('eating')
```

```python
reg.stem('ingeating')
```

# SNOWBALL STEMMING

```python
from nltk.stem import SnowballStemmer  
```

```python
stemmer = SnowballStemmer("english")
```

```python
for word in words:
    print(word + " -----> " + stemmer.stem(word)) 
```
